```python
import pandas as pd
from collections import Counter
from fare.metrics import rank_parity, rank_calibration, rank_equality
from fare.audit import audit_parity, audit_calibration, audit_equality, generate_diagnostics, plot_audit
```

```python
compasbaseline = pd.read_csv('compas_baseline_score.csv')
```

```python
compasbaseline
```

```python
def racebin(race):
    if race == 'Caucasian':
        return 'White'
    else:
        return 'NonWhite'
```

```python
compasbaseline['race_binary'] = compasbaseline['race'].apply(racebin)
```

```python
Counter(compasbaseline['race_binary'])
```

```python
race_factor= pd.get_dummies(compasbaseline['race_binary'],drop_first=True)
gender_factor = pd.get_dummies(compasbaseline['sex'],drop_first=True)
```

```python
compasbaseline = compasbaseline.drop(['race','sex','race_binary'],axis=1)
```

```python
compasbaseline = pd.concat([compasbaseline,race_factor,gender_factor],axis=1)
```
